```python
from gtts import gTTS
import IPython
```

```python
def text_to_speech(text):
    speech = gTTS(text = text, lang = "en", slow = False)
    speech.save("output.mp3")
    return IPython.display.Audio("output.mp3", autoplay=True)
```

```python
text = " Hi my name is Vijaykrishna. I am teaching you how to convert text to speech.  Can you try this. "
```

```python
text_to_speech(text)
```
